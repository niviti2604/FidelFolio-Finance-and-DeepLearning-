# Data Cleaning

# üìä Company Time-Series Imputation and Outlier Correction

## üìù Overview

This notebook performs **data imputation and correction** on a longitudinal company dataset. The dataset contains yearly feature values for multiple companies, some of which are **missing, inconsistent, or anomalous** due to reporting gaps or data entry errors. Our goal is to create a **complete, clean, and trustworthy dataset** for downstream analysis or modeling.

---

## üß™ Methodology

The process consists of the following steps:

### 1. **Temporal Expansion**
For each company, we expand the timeline from its **first available year** (assumed founding year) to **2020**, even if feature data is missing for intermediate years. This ensures consistency in the timeline.

---

### 2. **Spatial Founding Year Imputation**
If a company's feature values are missing in its founding year, we impute them using **spatial nearest neighbors**‚Äîi.e., similar companies in the same year based on feature similarity. This helps initialize missing sequences with realistic starting values.

---

### 3. **Temporal Imputation**
We apply **IterativeImputer** from `sklearn` for time-aware imputation. This approach models each feature as a function of the others and iteratively fills missing values. To enhance continuity over time, we also apply **linear interpolation** across the company's timeline.

---

### 4. **Spatial Outlier Correction**
We detect outliers in each year using **Isolation Forest**, a robust anomaly detection method. Identified outliers are corrected by replacing them with the **median of the three nearest non-outlier neighbors**, preserving the company‚Äôs contextual similarity.

---

### 5. **Final Validation and Export**
We validate that all missing values are imputed and outliers corrected. The cleaned dataset is exported as `Imputed_Complete_Company_Years.csv`, retaining the original target labels.

---

## ‚ö†Ô∏è Assumptions

- Each company's first valid year is treated as the founding year.
- Nearest neighbors are computed using standardized feature vectors.
- Outlier detection is done per year to maintain spatial consistency.

---

This pipeline is designed to be **modular, interpretable, and scalable**, suitable for both exploratory analysis and predictive modeling.


````python
# ======================================================================
# 1. Data Preparation & Temporal Expansion (Corrected)
# ======================================================================
def expand_temporal(df):
    """Ensure timeline from each company's founding year to 2020"""
    def expand_group(group):
        valid_years = group['Year'].dropna()
        if valid_years.empty:
            return group
        founding_year = int(valid_years.min())
        all_years = pd.DataFrame({'Year': range(founding_year, 2021)})
        merged = all_years.merge(group, how='left', on='Year')
        return merged
    
    return df.groupby('Company', group_keys=False).apply(expand_group).reset_index(drop=True)

# Load and clean data
data = pd.read_csv("FidelFolio_Dataset.csv")
data['Year'] = pd.to_numeric(data['Year'], errors='coerce')
data = expand_temporal(data)

feature_cols = [f'Feature{i}' for i in range(1, 29)]
for col in feature_cols:
    data[col] = pd.to_numeric(
        data[col].astype(str).str.replace('[^\\d.]', '', regex=True),
        errors='coerce'
    )

# ======================================================================
# 2. Spatial Founding Year Imputation
# ======================================================================
def impute_founding_year(group):
    """Impute missing values in founding year using spatial neighbors"""
    if group['Year'].isna().all():
        return group
    
    first_year = group['Year'].min()
    mask = group['Year'] == first_year
    
    if group.loc[mask, feature_cols].isna().all().all():
        yearly_data = data[data['Year'] == first_year].dropna(subset=feature_cols, how='all')
        if not yearly_data.empty:
            scaler = StandardScaler()
            scaled_data = scaler.fit_transform(yearly_data[feature_cols].fillna(0))
            nbrs = NearestNeighbors(n_neighbors=3).fit(scaled_data)
            distances, indices = nbrs.kneighbors(np.zeros((1, len(feature_cols))))
            group.loc[mask, feature_cols] = yearly_data.iloc[indices[0][0]][feature_cols].values
            
    return group

data = data.groupby('Company', group_keys=False).apply(impute_founding_year)

# ======================================================================
# 3. Temporal Imputation
# ======================================================================
def temporal_imputation(df):
    """Company-wise iterative imputation with temporal ordering"""
    df = df.sort_values('Year')
    
    valid_cols = df[feature_cols].columns[df[feature_cols].notna().any()]
    missing_cols = list(set(feature_cols) - set(valid_cols))
    df[missing_cols] = data[missing_cols].median()
    
    imputer = IterativeImputer(max_iter=20, random_state=42)
    df[valid_cols] = imputer.fit_transform(df[valid_cols])
    
    df[feature_cols] = df[feature_cols].interpolate(method='linear', limit_direction='both')
    return df

data = data.groupby('Company', group_keys=False).apply(temporal_imputation)

# ======================================================================
# 4. Spatial Outlier Correction
# ======================================================================
def correct_outliers(df):
    """Yearly spatial outlier detection and correction"""
    for year in df['Year'].unique():
        year_data = df[df['Year'] == year]
        if len(year_data) < 2:
            continue
            
        scaler = StandardScaler()
        scaled = scaler.fit_transform(year_data[feature_cols])
        iso = IsolationForest(contamination=0.05, random_state=42)
        outliers = iso.fit_predict(scaled)
        
        nbrs = NearestNeighbors(n_neighbors=3).fit(scaled)
        for idx in year_data.index[outliers == -1]:
            nn_indices = nbrs.kneighbors([scaled[year_data.index.get_loc(idx)]])[1][0]
            valid_nns = [i for i in nn_indices if outliers[i] == 1]
            if valid_nns:
                df.loc[idx, feature_cols] = year_data.iloc[valid_nns][feature_cols].median()
    
    return df

data = data.groupby('Year', group_keys=False).apply(correct_outliers)

# ======================================================================
# 5. Final Validation & Export
# ======================================================================
assert data[feature_cols].isna().sum().sum() == 0, "Remaining NaNs!"
target_cols = [' Target 1 ', ' Target 2 ', ' Target 3 ']
data[target_cols] = data[target_cols]

data.to_csv("Imputed_Complete_Company_Years.csv", index=False)
````

## üß™ Baseline Modeling: Naive and Global Mean Approaches

In this section, we evaluate simple baseline models to establish a reference point for future predictive modeling. These baselines help us assess how well advanced models perform compared to naive strategies.

---

### üßº Data Cleaning

- Strip any leading/trailing whitespace from column names.
- Drop rows with missing values in the `Target 1` column.
- Ensure `Target 1` is numeric by coercing non-numeric values to NaN and dropping them.

---

- **Training Set**: All data up to and including the year **2015**.
- **Testing Set**: All data after the year **2015**.

This temporal split ensures we predict future values without data leakage.

---

### üìâ Naive Baseline: Last Observed Value

We use a **last-observed-value** strategy for each company as a simple forecast:

- For each company, take the **last available `Target 1` value** from the training set.
- Use this value as the prediction for all future years (in the test set).

We then compute the **Root Mean Squared Error (RMSE)** to evaluate performance.

### üîÑ Global Mean Baseline

We use the **global mean** of `Target 1` in the training set as a simple forecast:

- **Calculate the mean** of `Target 1` from the training set.
- **Use this value** as the prediction for all years in the test set.

We then compute the **Root Mean Squared Error (RMSE)** to evaluate performance.

By comparing the **RMSE of the Naive Baseline** and **Global Mean Baseline**, we get an initial understanding of how well the models perform relative to the simplest strategies.


````python
# Clean data
data.columns = data.columns.str.strip()
data = data.dropna(subset=['Target 1'])
data['Target 1'] = pd.to_numeric(data['Target 1'], errors='coerce')
data = data.dropna(subset=['Target 1'])

# Train/test split
train = data[data['Year'] <= 2015]
test = data[data['Year'] > 2015]

# Naive baseline: Last observed value per company
last_values = train.groupby('Company')['Target 1'].last().to_dict()
test['naive_pred'] = test['Company'].map(last_values)

# Drop rows with NaN predictions (companies not in train set)
test_clean = test.dropna(subset=['naive_pred'])
rmse_naive = sqrt(mean_squared_error(test_clean['Target 1'], test_clean['naive_pred']))
print(f"Naive Baseline RMSE: {rmse_naive}")

# Optional: Compare to global mean baseline
global_mean = train['Target 1'].mean()
test['global_mean_pred'] = global_mean
rmse_global_mean = sqrt(mean_squared_error(test['Target 1'], test['global_mean_pred']))
print(f"Global Mean Baseline RMSE: {rmse_global_mean}")
````

## üîÅ Time-Series Sequence Preparation for Modeling

To enable sequence-based learning models (such as RNNs, LSTMs, or 1D CNNs), we convert the company-wise temporal data into **sliding window sequences**. This process transforms the dataset into samples of fixed-length historical windows paired with the next year's target value.

---

### üßº Preprocessing Steps

1. **Drop Null Targets**  
   Ensure all rows used for modeling have valid `Target 1` values.

2. **Sort Chronologically**  
   Sort data by `Company` and `Year` to preserve correct temporal order for each company.

---

### üß© Sequence Generation

We use a **sliding window** approach to generate training samples:

- For each company:
  - Extract rolling windows of `Feature1` to `Feature28` for the past `N` years.
  - Predict the `Target 1` value of the **next year**.
- The `window_size` controls how many past years are used for prediction (e.g., 5 years).

````python
# Drop rows where Target 1 is null
data = data.dropna(subset=['Target 1'])

# Sort data by Company and Year
data = data.sort_values(['Company', 'Year'])

# Group by Company to create time-series sequences
def create_sequences(group, window_size=3):
    features = group.filter(like='Feature').values
    target = group['Target 1'].values
    sequences = []
    for i in range(len(features) - window_size):
        seq = features[i:i+window_size]
        label = target[i+window_size]
        sequences.append((seq, label))
    return sequences

window_size = 5  # Adjust based on your needs
sequences = []
for _, group in data.groupby('Company'):
    if len(group) > window_size:
        sequences += create_sequences(group, window_size)

# Split into features and labels
X = np.array([s[0] for s in sequences])
y = np.array([s[1] for s in sequences])

# Train-test split (time-series friendly)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Scale features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.reshape(-1, 28)).reshape(X_train.shape)
X_test = scaler.transform(X_test.reshape(-1, 28)).reshape(X_test.shape)
````

## üß† GRU-Based Sequence Modeling with Hyperparameter Tuning

This section develops a **GRU (Gated Recurrent Unit)** model to predict `Target 1` using past company data represented as time-series sequences.

---

### üóÉÔ∏è Data Preparation

1. **Cleaning**:  
   Remove rows where `Target 1` is missing and ensure it's numeric.

2. **Time-Based Split**:  
   Divide the dataset into:
   - **Training set**: Years ‚â§ 2015
   - **Testing set**: Years > 2015

---

### üîÅ Sequence Generation

- Use a **sliding window** of `window_size` years for each company.
- For training:
  - Generate all possible rolling windows of features.
- For testing:
  - Use the **last `window_size` years** from training to predict the **first available test year** for each company.

This ensures test predictions are made **without peeking into the future**.

---

### ‚öñÔ∏è Feature Scaling

Apply `StandardScaler` to standardize features:

- Fit on training set.
- Transform both training and testing sets.

Reshaping ensures proper scaling in 3D (samples √ó timesteps √ó features).

---

### üß™ Model Architecture & Hyperparameter Tuning

We use **Keras Tuner** for random search over GRU architecture:

- **GRU Layer**: Units between 32‚Äì128  
- **Dropout**: Between 0.1‚Äì0.5 to prevent overfitting  
- **Learning Rate**: Try different values for `Adam` optimizer

````python
# Load and prepare data
data.columns = data.columns.str.strip()
data = data.dropna(subset=['Target 1'])
data['Target 1'] = pd.to_numeric(data['Target 1'], errors='coerce')
data = data.dropna(subset=['Target 1'])

# Time-based split
train = data[data['Year'] <= 2015]
test = data[data['Year'] > 2015]

# Feature columns (adjust based on your actual features)
feature_cols = [col for col in data.columns if 'Feature' in col]

def create_sequences(df, window_size=3):
    sequences = []
    targets = []
    for company, group in df.groupby('Company'):
        company_data = group.sort_values('Year')
        if len(company_data) < window_size + 1:
            continue
        features = company_data[feature_cols].values
        target = company_data['Target 1'].values
        for i in range(len(features) - window_size):
            sequences.append(features[i:i+window_size])
            targets.append(target[i+window_size])
    return np.array(sequences), np.array(targets)

# Create sequences from training data
window_size = 3
X_train, y_train = create_sequences(train, window_size)

# Create test sequences (only first prediction per company)
X_test, y_test = [], []
for company, group in test.groupby('Company'):
    company_train = train[train['Company'] == company]
    if len(company_train) >= window_size:
        seq = company_train.sort_values('Year')[feature_cols].values[-window_size:]
        target = group.sort_values('Year')['Target 1'].values[0]
        X_test.append(seq)
        y_test.append(target)
X_test, y_test = np.array(X_test), np.array(y_test)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.reshape(-1, len(feature_cols))).reshape(X_train.shape)
X_test = scaler.transform(X_test.reshape(-1, len(feature_cols))).reshape(X_test.shape)

# Simplified model architecture
def build_model(hp):
    model = Sequential()
    model.add(GRU(
        units=hp.Int('units', 32, 128, step=32),
        input_shape=(window_size, len(feature_cols)),
        return_sequences=False
    ))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# Hyperparameter tuning
tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='gru_tuning'
)

early_stop = EarlyStopping(patience=5, restore_best_weights=True)

tuner.search(X_train, y_train,
             epochs=100,
             validation_split=0.2,
             callbacks=[early_stop],
             verbose=1)

# Get best model
best_model = tuner.get_best_models(num_models=1)[0]

# Final training
history = best_model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Evaluation
test_pred = best_model.predict(X_test)
test_rmse = sqrt(mean_squared_error(y_test, test_pred))
print(f"Test RMSE: {test_rmse:.2f}")
````

```python

````python
# Load and prepare data
data.columns = data.columns.str.strip()
data = data.dropna(subset=['Target 1'])
data['Target 1'] = pd.to_numeric(data['Target 1'], errors='coerce')
data = data.dropna(subset=['Target 1'])

# Time-based split
train = data[data['Year'] <= 2015]
test = data[data['Year'] > 2015]

# Feature columns (adjust based on your actual features)
feature_cols = [col for col in data.columns if 'Feature' in col]

def create_sequences(df, window_size=3):
    sequences = []
    targets = []
    for company, group in df.groupby('Company'):
        company_data = group.sort_values('Year')
        if len(company_data) < window_size + 1:
            continue
        features = company_data[feature_cols].values
        target = company_data['Target 1'].values
        for i in range(len(features) - window_size):
            sequences.append(features[i:i+window_size])
            targets.append(target[i+window_size])
    return np.array(sequences), np.array(targets)

# Create sequences from training data
window_size = 3
X_train, y_train = create_sequences(train, window_size)

# Create test sequences (only first prediction per company)
X_test, y_test = [], []
for company, group in test.groupby('Company'):
    company_train = train[train['Company'] == company]
    if len(company_train) >= window_size:
        seq = company_train.sort_values('Year')[feature_cols].values[-window_size:]
        target = group.sort_values('Year')['Target 1'].values[0]
        X_test.append(seq)
        y_test.append(target)
X_test, y_test = np.array(X_test), np.array(y_test)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.reshape(-1, len(feature_cols))).reshape(X_train.shape)
X_test_scaled = scaler.transform(X_test.reshape(-1, len(feature_cols))).reshape(X_test.shape)

# 1. LSTM Model
def build_lstm_model(hp):
    model = Sequential()
    model.add(LSTM(
        units=hp.Int('units', 32, 256, step=32),
        input_shape=(window_size, len(feature_cols)),
        return_sequences=False
    ))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# 2. CNN-LSTM Model
def build_cnn_lstm_model(hp):
    model = Sequential()
    # CNN part
    model.add(TimeDistributed(Conv1D(
        filters=hp.Int('filters', 32, 128, step=32),
        kernel_size=hp.Int('kernel_size', 2, 3),
        activation='relu'),
        input_shape=(window_size, len(feature_cols), 1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=1)))
    model.add(TimeDistributed(Flatten()))
    
    # LSTM part
    model.add(LSTM(
        units=hp.Int('lstm_units', 32, 128, step=32),
        return_sequences=False))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# Reshape data for CNN-LSTM (add channel dimension)
X_train_cnn = X_train_scaled.reshape(X_train_scaled.shape[0], window_size, len(feature_cols), 1)
X_test_cnn = X_test_scaled.reshape(X_test_scaled.shape[0], window_size, len(feature_cols), 1)

# Tune LSTM model
lstm_tuner = kt.RandomSearch(
    build_lstm_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='lstm_tuning'
)

# Tune CNN-LSTM model
cnn_lstm_tuner = kt.RandomSearch(
    build_cnn_lstm_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='cnn_lstm_tuning'
)

early_stop = EarlyStopping(patience=5, restore_best_weights=True)

print("Tuning LSTM model...")
lstm_tuner.search(X_train_scaled, y_train,
                 epochs=100,
                 validation_split=0.2,
                 callbacks=[early_stop],
                 verbose=1)

print("\nTuning CNN-LSTM model...")
cnn_lstm_tuner.search(X_train_cnn, y_train,
                     epochs=100,
                     validation_split=0.2,
                     callbacks=[early_stop],
                     verbose=1)

# Get best models
best_lstm = lstm_tuner.get_best_models(num_models=1)[0]
best_cnn_lstm = cnn_lstm_tuner.get_best_models(num_models=1)[0]

# Train final models
print("\nTraining final LSTM model...")
lstm_history = best_lstm.fit(
    X_train_scaled, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

print("\nTraining final CNN-LSTM model...")
cnn_lstm_history = best_cnn_lstm.fit(
    X_train_cnn, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Evaluate models
def evaluate_model(model, X_test_data):
    pred = model.predict(X_test_data)
    rmse = sqrt(mean_squared_error(y_test, pred))
    return rmse

lstm_rmse = evaluate_model(best_lstm, X_test_scaled)
cnn_lstm_rmse = evaluate_model(best_cnn_lstm, X_test_cnn)

print("\nModel Comparison:")
print(f"LSTM Test RMSE: {lstm_rmse:.2f}")
print(f"CNN-LSTM Test RMSE: {cnn_lstm_rmse:.2f}")

# Save best model
if lstm_rmse < cnn_lstm_rmse:
    best_model = best_lstm
    print("\nSelected LSTM as the best model")
else:
    best_model = best_cnn_lstm
    print("\nSelected CNN-LSTM as the best model")
````

## üß™ Model Training and Evaluation

In this section, we explain the mathematical aspects of data preparation, sequence generation, feature scaling, model architecture, hyperparameter tuning, and evaluation.

---

### üßº Data Preparation

The data is cleaned by:
- **Removing missing values** in the `Target 2` column.
- **Converting** non-numeric entries in `Target 2` to `NaN` and then removing them.

The data is split into two sets:
- **Training Set**: All data where the year is less than or equal to 2015.
- **Testing Set**: All data where the year is greater than 2015.
  
This ensures that the model is evaluated on data it has not seen during training, avoiding data leakage.

---

### üîÄ Sequence Generation

For each company, we generate sequences of feature values and their corresponding target values. The key steps are:

1. **Feature Sequences**: For each company, the data is sorted by year. We then form sequences of features of length `window_size` (in this case, 3), which represent the past few time steps.
2. **Target Values**: For each sequence, the target value is the `Target 2` value at the time step immediately following the sequence.

For example, if `window_size = 3`, the model will use the first three data points of features to predict the fourth data point's target value.

---

### üìä Feature Scaling

The features are **standardized** using a scaling technique:
- The **StandardScaler** standardizes each feature to have a **mean of 0** and a **standard deviation of 1**.

This is calculated by subtracting the feature mean and dividing by its standard deviation:
- For each feature `x_i`, the scaled value is computed as:

  `x_i_scaled = (x_i - mean_x) / std_x`

Where:
- `mean_x` is the mean of the feature `x`,
- `std_x` is the standard deviation of the feature `x`.

This ensures that the model learns efficiently and is not biased by the scale of the input features.

---

### üèóÔ∏è Model Architecture

The model is based on a **GRU (Gated Recurrent Unit)** network, which is a type of recurrent neural network (RNN). The key elements are:

1. **GRU Layer**: The GRU unit is designed to capture the temporal dependencies in the data by maintaining a hidden state that is updated over time. The update rule for the GRU is as follows:

  `h_t = (1 - z_t) * h_(t-1) + z_t * h_tilde`

Where:
- `h_t` is the hidden state at time step `t`,
- `z_t` is the update gate (how much of the previous state to keep),
- `h_tilde` is the candidate hidden state, and
- `*` represents element-wise multiplication.

2. **Dropout Layer**: Dropout is used to prevent overfitting by randomly setting a fraction of the inputs to zero during training. This regularization technique helps the model generalize better.

3. **Dense Layer**: A fully connected layer that outputs a single value, which is the predicted target for each sequence.

The model is trained to minimize the **Mean Squared Error (MSE)** loss function, which is calculated as:

  `MSE = (1 / N) * Œ£(y_i - y_pred_i)^2`

Where:
- `y_i` is the true target value,
- `y_pred_i` is the predicted target value, and
- `N` is the number of samples.

---

### üîß Hyperparameter Tuning

We use **Random Search** to tune several hyperparameters:
- **Number of units** in the GRU layer: This controls the model's capacity to learn temporal patterns.
- **Dropout rate**: This controls the probability of setting an input to zero in the dropout layer.
- **Learning rate**: The rate at which the model adjusts its weights during training.

The goal of hyperparameter tuning is to find the best combination of these parameters that minimizes the validation loss (MSE).

---

### üèãÔ∏è‚Äç‚ôÇÔ∏è Model Training

During training, the model is optimized using the **Adam optimizer**. The update rule for the Adam optimizer is as follows:

  `Œ∏_t = Œ∏_(t-1) - Œ∑ * (m_t / ‚àö(v_t) + Œµ)`

Where:
- `Œ∏_t` is the model parameters at time step `t`,
- `Œ∑` is the learning rate,
- `m_t` is the first moment estimate (mean of the gradients),
- `v_t` is the second moment estimate (variance of the gradients),
- `Œµ` is a small constant to prevent division by zero.

The model is trained until the validation loss stops improving, using **early stopping** to avoid overfitting.

---

### üìà Model Evaluation

Finally, the model's performance is evaluated on the test set using **Root Mean Squared Error (RMSE)**:

  `RMSE = ‚àö((1 / N) * Œ£(y_i - y_pred_i)^2)`

Where:
- `y_i` is the true target value,
- `y_pred_i` is the predicted target value, and
- `N` is the number of test samples.

RMSE gives an indication of how well the model is predicting the target values, with lower RMSE values indicating better performance.

---


````python
# Load and prepare data
data.columns = data.columns.str.strip()
data = data.dropna(subset=['Target 2'])
data['Target 2'] = pd.to_numeric(data['Target 2'], errors='coerce')
data = data.dropna(subset=['Target 2'])

# Time-based split
train = data[data['Year'] <= 2015]
test = data[data['Year'] > 2015]

# Feature columns (adjust based on your actual features)
feature_cols = [col for col in data.columns if 'Feature' in col]

def create_sequences(df, window_size=3):
    sequences = []
    targets = []
    for company, group in df.groupby('Company'):
        company_data = group.sort_values('Year')
        if len(company_data) < window_size + 1:
            continue
        features = company_data[feature_cols].values
        target = company_data['Target 2'].values
        for i in range(len(features) - window_size):
            sequences.append(features[i:i+window_size])
            targets.append(target[i+window_size])
    return np.array(sequences), np.array(targets)

# Create sequences from training data
window_size = 3
X_train, y_train = create_sequences(train, window_size)

# Create test sequences (only first prediction per company)
X_test, y_test = [], []
for company, group in test.groupby('Company'):
    company_train = train[train['Company'] == company]
    if len(company_train) >= window_size:
        seq = company_train.sort_values('Year')[feature_cols].values[-window_size:]
        target = group.sort_values('Year')['Target 2'].values[0]
        X_test.append(seq)
        y_test.append(target)
X_test, y_test = np.array(X_test), np.array(y_test)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.reshape(-1, len(feature_cols))).reshape(X_train.shape)
X_test = scaler.transform(X_test.reshape(-1, len(feature_cols))).reshape(X_test.shape)

# Simplified model architecture
def build_model(hp):
    model = Sequential()
    model.add(GRU(
        units=hp.Int('units', 32, 128, step=32),
        input_shape=(window_size, len(feature_cols)),
        return_sequences=False
    ))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# Hyperparameter tuning
tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='gru_tuning'
)

early_stop = EarlyStopping(patience=5, restore_best_weights=True)

tuner.search(X_train, y_train,
             epochs=100,
             validation_split=0.2,
             callbacks=[early_stop],
             verbose=1)

# Get best model
best_model = tuner.get_best_models(num_models=1)[0]

# Final training
history = best_model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Evaluation
test_pred = best_model.predict(X_test)
test_rmse = sqrt(mean_squared_error(y_test, test_pred))
print(f"Test RMSE: {test_rmse:.2f}")
````

3. **Dropout Layer**: Again used to prevent overfitting.
4. **Dense Layer**: Outputs a single value for the target.

### üèãÔ∏è‚Äç‚ôÇÔ∏è Model Training

- `y_i` is the actual target value for the i-th data point,

### üèÜ Model Comparison

---

### üóÇÔ∏è Model Selection

After evaluation, the model with the lower RMSE (either LSTM or CNN-LSTM) is selected as the best model for future predictions.


````python
# Load and prepare data
data.columns = data.columns.str.strip()
data = data.dropna(subset=['Target 2'])
data['Target 2'] = pd.to_numeric(data['Target 2'], errors='coerce')
data = data.dropna(subset=['Target 2'])

# Time-based split
train = data[data['Year'] <= 2015]
test = data[data['Year'] > 2015]

# Feature columns (adjust based on your actual features)
feature_cols = [col for col in data.columns if 'Feature' in col]

def create_sequences(df, window_size=3):
    sequences = []
    targets = []
    for company, group in df.groupby('Company'):
        company_data = group.sort_values('Year')
        if len(company_data) < window_size + 1:
            continue
        features = company_data[feature_cols].values
        target = company_data['Target 2'].values
        for i in range(len(features) - window_size):
            sequences.append(features[i:i+window_size])
            targets.append(target[i+window_size])
    return np.array(sequences), np.array(targets)

# Create sequences from training data
window_size = 3
X_train, y_train = create_sequences(train, window_size)

# Create test sequences (only first prediction per company)
X_test, y_test = [], []
for company, group in test.groupby('Company'):
    company_train = train[train['Company'] == company]
    if len(company_train) >= window_size:
        seq = company_train.sort_values('Year')[feature_cols].values[-window_size:]
        target = group.sort_values('Year')['Target 2'].values[0]
        X_test.append(seq)
        y_test.append(target)
X_test, y_test = np.array(X_test), np.array(y_test)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.reshape(-1, len(feature_cols))).reshape(X_train.shape)
X_test_scaled = scaler.transform(X_test.reshape(-1, len(feature_cols))).reshape(X_test.shape)

# 1. LSTM Model
def build_lstm_model(hp):
    model = Sequential()
    model.add(LSTM(
        units=hp.Int('units', 32, 256, step=32),
        input_shape=(window_size, len(feature_cols)),
        return_sequences=False
    ))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# 2. CNN-LSTM Model
def build_cnn_lstm_model(hp):
    model = Sequential()
    # CNN part
    model.add(TimeDistributed(Conv1D(
        filters=hp.Int('filters', 32, 128, step=32),
        kernel_size=hp.Int('kernel_size', 2, 3),
        activation='relu'),
        input_shape=(window_size, len(feature_cols), 1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=1)))
    model.add(TimeDistributed(Flatten()))
    
    # LSTM part
    model.add(LSTM(
        units=hp.Int('lstm_units', 32, 128, step=32),
        return_sequences=False))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# Reshape data for CNN-LSTM (add channel dimension)
X_train_cnn = X_train_scaled.reshape(X_train_scaled.shape[0], window_size, len(feature_cols), 1)
X_test_cnn = X_test_scaled.reshape(X_test_scaled.shape[0], window_size, len(feature_cols), 1)

# Tune LSTM model
lstm_tuner = kt.RandomSearch(
    build_lstm_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='lstm_tuning'
)

# Tune CNN-LSTM model
cnn_lstm_tuner = kt.RandomSearch(
    build_cnn_lstm_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='cnn_lstm_tuning'
)

early_stop = EarlyStopping(patience=5, restore_best_weights=True)

print("Tuning LSTM model...")
lstm_tuner.search(X_train_scaled, y_train,
                 epochs=100,
                 validation_split=0.2,
                 callbacks=[early_stop],
                 verbose=1)

print("\nTuning CNN-LSTM model...")
cnn_lstm_tuner.search(X_train_cnn, y_train,
                     epochs=100,
                     validation_split=0.2,
                     callbacks=[early_stop],
                     verbose=1)

# Get best models
best_lstm = lstm_tuner.get_best_models(num_models=1)[0]
best_cnn_lstm = cnn_lstm_tuner.get_best_models(num_models=1)[0]

# Train final models
print("\nTraining final LSTM model...")
lstm_history = best_lstm.fit(
    X_train_scaled, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

print("\nTraining final CNN-LSTM model...")
cnn_lstm_history = best_cnn_lstm.fit(
    X_train_cnn, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Evaluate models
def evaluate_model(model, X_test_data):
    pred = model.predict(X_test_data)
    rmse = sqrt(mean_squared_error(y_test, pred))
    return rmse

lstm_rmse = evaluate_model(best_lstm, X_test_scaled)
cnn_lstm_rmse = evaluate_model(best_cnn_lstm, X_test_cnn)

print("\nModel Comparison:")
print(f"LSTM Test RMSE: {lstm_rmse:.2f}")
print(f"CNN-LSTM Test RMSE: {cnn_lstm_rmse:.2f}")

# Save best model
if lstm_rmse < cnn_lstm_rmse:
    best_model = best_lstm
    print("\nSelected LSTM as the best model")
else:
    best_model = best_cnn_lstm
    print("\nSelected CNN-LSTM as the best model")
````

### 1. **Decision Tree Regressor**

#### Hyperparameter Tuning:
- The **GridSearchCV** is used to tune the following hyperparameters:

In this section, we train a **GRU (Gated Recurrent Unit)** model to predict `Target 3` based on the features. We use a time-series approach with sequences of previous observations to predict future values. The process includes:

4. **Feature Scaling**:
   - Scale the feature data using `StandardScaler` to ensure all features are on the same scale.

---

### 1. **GRU Model Architecture**

The **GRU model** is a type of Recurrent Neural Network (RNN) that is designed to handle sequences of data. In this model, we use:

- **GRU Layer**: A recurrent layer with units that can capture temporal dependencies.
- **Dropout Layer**: To avoid overfitting by randomly dropping connections during training.
- **Dense Layer**: The final layer to output a single prediction value for `Target 3`.

The model is compiled with the **Adam optimizer** and **Mean Squared Error (MSE)** as the loss function.

### 2. **Hyperparameter Tuning**

We use **Random Search** to tune the following hyperparameters:

- **Units** in the GRU layer: The number of units in the GRU layer, ranging from 32 to 128.
- **Dropout rate**: The rate of dropout, ranging from 0.1 to 0.5.
- **Learning rate**: The learning rate for the Adam optimizer, with values of `1e-3`, `5e-4`, and `1e-4`.

The **early stopping** callback is used to stop training if the validation loss does not improve for 5 consecutive epochs.

### 3. **Training the Model**

The **best model** from the hyperparameter tuning process is selected and trained on the full training data for 200 epochs, with a batch size of 32.

### 4. **Evaluation**

After training, the model is evaluated on the test set, and the **Root Mean Squared Error (RMSE)** is calculated to assess the model's performance.

The RMSE is calculated as:

`RMSE = ‚àö((1 / N) * Œ£(y_i - y_pred_i)^2)`

Where:
- `y_i` is the actual target value for the i-th data point,
- `y_pred_i` is the predicted target value for the i-th data point,
- `N` is the number of data points in the test set.



````python
# Load and prepare data
data.columns = data.columns.str.strip()
data = data.dropna(subset=['Target 3'])
data['Target 3'] = pd.to_numeric(data['Target 3'], errors='coerce')
data = data.dropna(subset=['Target 3'])

# Time-based split
train = data[data['Year'] <= 2015]
test = data[data['Year'] > 2015]

# Feature columns (adjust based on your actual features)
feature_cols = [col for col in data.columns if 'Feature' in col]

def create_sequences(df, window_size=3):
    sequences = []
    targets = []
    for company, group in df.groupby('Company'):
        company_data = group.sort_values('Year')
        if len(company_data) < window_size + 1:
            continue
        features = company_data[feature_cols].values
        target = company_data['Target 3'].values
        for i in range(len(features) - window_size):
            sequences.append(features[i:i+window_size])
            targets.append(target[i+window_size])
    return np.array(sequences), np.array(targets)

# Create sequences from training data
window_size = 3
X_train, y_train = create_sequences(train, window_size)

# Create test sequences (only first prediction per company)
X_test, y_test = [], []
for company, group in test.groupby('Company'):
    company_train = train[train['Company'] == company]
    if len(company_train) >= window_size:
        seq = company_train.sort_values('Year')[feature_cols].values[-window_size:]
        target = group.sort_values('Year')['Target 3'].values[0]
        X_test.append(seq)
        y_test.append(target)
X_test, y_test = np.array(X_test), np.array(y_test)

# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train.reshape(-1, len(feature_cols))).reshape(X_train.shape)
X_test = scaler.transform(X_test.reshape(-1, len(feature_cols))).reshape(X_test.shape)

# Simplified model architecture
def build_model(hp):
    model = Sequential()
    model.add(GRU(
        units=hp.Int('units', 32, 128, step=32),
        input_shape=(window_size, len(feature_cols)),
        return_sequences=False
    ))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# Hyperparameter tuning
tuner = kt.RandomSearch(
    build_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='gru_tuning'
)

early_stop = EarlyStopping(patience=5, restore_best_weights=True)

tuner.search(X_train, y_train,
             epochs=100,
             validation_split=0.2,
             callbacks=[early_stop],
             verbose=1)

# Get best model
best_model = tuner.get_best_models(num_models=1)[0]

# Final training
history = best_model.fit(
    X_train, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Evaluation
test_pred = best_model.predict(X_test)
test_rmse = sqrt(mean_squared_error(y_test, test_pred))
print(f"Test RMSE: {test_rmse:.2f}")
````

RMSE = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}
\]

Where:
- \( y_i \) is the actual target value for the \( i^{th} \) data point,
- \( \hat{y}_i \) is the predicted value for the \( i^{th} \) data point,

````python
# Load and prepare data
data.columns = data.columns.str.strip()
data = data.dropna(subset=['Target 3'])
data['Target 3'] = pd.to_numeric(data['Target 3'], errors='coerce')
data = data.dropna(subset=['Target 3'])

# Time-based split
train = data[data['Year'] <= 2015]
test = data[data['Year'] > 2015]

# Feature columns (adjust based on your actual features)
feature_cols = [col for col in data.columns if 'Feature' in col]

def create_sequences(df, window_size=3):
    sequences = []
    targets = []
    for company, group in df.groupby('Company'):
        company_data = group.sort_values('Year')
        if len(company_data) < window_size + 1:
            continue
        features = company_data[feature_cols].values
        target = company_data['Target 3'].values
        for i in range(len(features) - window_size):
            sequences.append(features[i:i+window_size])
            targets.append(target[i+window_size])
    return np.array(sequences), np.array(targets)

# Create sequences from training data
window_size = 3
X_train, y_train = create_sequences(train, window_size)

# Create test sequences (only first prediction per company)
X_test, y_test = [], []
for company, group in test.groupby('Company'):
    company_train = train[train['Company'] == company]
    if len(company_train) >= window_size:
        seq = company_train.sort_values('Year')[feature_cols].values[-window_size:]
        target = group.sort_values('Year')['Target 3'].values[0]
        X_test.append(seq)
        y_test.append(target)
X_test, y_test = np.array(X_test), np.array(y_test)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.reshape(-1, len(feature_cols))).reshape(X_train.shape)
X_test_scaled = scaler.transform(X_test.reshape(-1, len(feature_cols))).reshape(X_test.shape)

# 1. LSTM Model
def build_lstm_model(hp):
    model = Sequential()
    model.add(LSTM(
        units=hp.Int('units', 32, 256, step=32),
        input_shape=(window_size, len(feature_cols)),
        return_sequences=False
    ))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# 2. CNN-LSTM Model
def build_cnn_lstm_model(hp):
    model = Sequential()
    # CNN part
    model.add(TimeDistributed(Conv1D(
        filters=hp.Int('filters', 32, 128, step=32),
        kernel_size=hp.Int('kernel_size', 2, 3),
        activation='relu'),
        input_shape=(window_size, len(feature_cols), 1)))
    model.add(TimeDistributed(MaxPooling1D(pool_size=1)))
    model.add(TimeDistributed(Flatten()))
    
    # LSTM part
    model.add(LSTM(
        units=hp.Int('lstm_units', 32, 128, step=32),
        return_sequences=False))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(1))
    
    model.compile(
        optimizer=Adam(hp.Choice('learning_rate', [1e-3, 5e-4, 1e-4])),
        loss='mse'
    )
    return model

# Reshape data for CNN-LSTM (add channel dimension)
X_train_cnn = X_train_scaled.reshape(X_train_scaled.shape[0], window_size, len(feature_cols), 1)
X_test_cnn = X_test_scaled.reshape(X_test_scaled.shape[0], window_size, len(feature_cols), 1)

# Tune LSTM model
lstm_tuner = kt.RandomSearch(
    build_lstm_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='lstm_tuning'
)

# Tune CNN-LSTM model
cnn_lstm_tuner = kt.RandomSearch(
    build_cnn_lstm_model,
    objective='val_loss',
    max_trials=10,
    directory='tuning',
    project_name='cnn_lstm_tuning'
)

early_stop = EarlyStopping(patience=5, restore_best_weights=True)

print("Tuning LSTM model...")
lstm_tuner.search(X_train_scaled, y_train,
                 epochs=100,
                 validation_split=0.2,
                 callbacks=[early_stop],
                 verbose=1)

print("\nTuning CNN-LSTM model...")
cnn_lstm_tuner.search(X_train_cnn, y_train,
                     epochs=100,
                     validation_split=0.2,
                     callbacks=[early_stop],
                     verbose=1)

# Get best models
best_lstm = lstm_tuner.get_best_models(num_models=1)[0]
best_cnn_lstm = cnn_lstm_tuner.get_best_models(num_models=1)[0]

# Train final models
print("\nTraining final LSTM model...")
lstm_history = best_lstm.fit(
    X_train_scaled, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

print("\nTraining final CNN-LSTM model...")
cnn_lstm_history = best_cnn_lstm.fit(
    X_train_cnn, y_train,
    epochs=200,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)

# Evaluate models
def evaluate_model(model, X_test_data):
    pred = model.predict(X_test_data)
    rmse = sqrt(mean_squared_error(y_test, pred))
    return rmse

lstm_rmse = evaluate_model(best_lstm, X_test_scaled)
cnn_lstm_rmse = evaluate_model(best_cnn_lstm, X_test_cnn)

print("\nModel Comparison:")
print(f"LSTM Test RMSE: {lstm_rmse:.2f}")
print(f"CNN-LSTM Test RMSE: {cnn_lstm_rmse:.2f}")

# Save best model
if lstm_rmse < cnn_lstm_rmse:
    best_model = best_lstm
    print("\nSelected LSTM as the best model")
else:
    best_model = best_cnn_lstm
    print("\nSelected CNN-LSTM as the best model")
````

     - Number of **LSTM units** (between 32 and 256).
     - Number of **dense units** and **activation functions** for the fully connected layer.
     - Learning rate for the **Adam optimizer**.
   - **Early Stopping**: The model uses **early stopping** with a patience of 10 epochs to avoid overfitting.
   
   - After hyperparameter tuning, the best LSTM model is obtained.

### 2. **Feature Extraction**
   - **Feature Extractor**: The penultimate layer of the LSTM model is used to extract features.
   - **Feature Prediction**: Features are extracted for both the training and test datasets using the **feature extractor**.

### 3. **Data Validation**
   - **Dimension Validation**: A set of checks is added before tuning to ensure that the input data dimensions match the expected ones:
     - **Expected Shape**: `(TIMESTEPS, N_FEATURES)`.

   - **Assertions**: Checks confirm the timesteps and feature dimensions are correct.

### 4. **Support Vector Regressor (SVR) Hyperparameter Tuning**
   - **SVR Hyperparameters**: A **RandomizedSearchCV** is used to tune the hyperparameters of the **SVR**:
     - **C** (Regularization parameter).
     - **epsilon** (Margin of tolerance).
     - **kernel** type (RBF, polynomial, and sigmoid).
   
   - **Feature Scaling**: The features extracted by the LSTM are scaled using **StandardScaler** before feeding them into the SVR.

   - After tuning, the best **SVR** model is selected.

### 5. **Final Ensemble Model**
   - **Ensemble Model**: The final model combines both LSTM and SVR:
     - The **LSTM model** is used to extract features.
     - The extracted features are scaled and passed to the **SVR model** for prediction.
   
   - **Model Evaluation**: The performance of the ensemble model is evaluated using **RMSE**.

---

### 6. **Results**
After training and evaluation, the **RMSE** is calculated for the final ensemble model, and the best hyperparameters for both the LSTM and SVR are printed:

\[
RMSE = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}
\]

Where:
- \( y_i \) is the actual target value for the \( i^{th} \) data point,
- \( \hat{y}_i \) is the predicted value for the \( i^{th} \) data point,
- \( N \) is the number of data points.

The **best LSTM parameters** and **best SVR parameters** are also printed to show the optimal configuration for both models.



````python
# --------- Corrected Hyperparameters ---------
# Get dimensions from actual training data
TIMESTEPS = X_train.shape[1]  # Should be 3 based on error message
N_FEATURES = X_train.shape[2]  # Should be 28 based on error
LSTM_EPOCHS = 100
SVR_CV = 5

# --------- Updated LSTM Model Building ---------
def build_lstm(hp):
    inputs = Input(shape=(TIMESTEPS, N_FEATURES))  # Match actual input dimensions
    x = LSTM(
        units=hp.Int('units', 32, 256, step=32),
        return_sequences=False
    )(inputs)
    x = Dense(
        hp.Int('dense_units', 16, 128, step=16), 
        activation=hp.Choice('dense_act', ['relu', 'tanh'])
    )(x)
    outputs = Dense(1)(x)
    
    model = Model(inputs=inputs, outputs=outputs)
    model.compile(
        optimizer=Adam(hp.Float('lr', 1e-4, 1e-2, sampling='log')),
        loss='mean_squared_error'
    )
    return model

lstm_tuner = kt.RandomSearch(
    build_lstm,
    objective='val_loss',
    max_trials=20,
    directory='lstm_tuning',
    project_name='lstm_regression'
)

early_stop = EarlyStopping(patience=10, restore_best_weights=True)

# Assuming X_train, X_val, y_train, y_val from your data
lstm_tuner.search(
    X_train, y_train,
    epochs=LSTM_EPOCHS,
    validation_split=0.2,
    callbacks=[early_stop],
    batch_size=64,
    verbose=1
)

# Get best LSTM model
best_lstm = lstm_tuner.get_best_models(num_models=1)[0]

# --------- 2. Feature Extraction ---------
feature_extractor = Model(
    inputs=best_lstm.input,
    outputs=best_lstm.layers[-2].output  # Get penultimate layer outputs
)

# Extract features
X_train_features = feature_extractor.predict(X_train)
X_test_features = feature_extractor.predict(X_test)

# --------- Data Validation Steps ---------
# Add these checks before tuning
print(f"Training data shape: {X_train.shape}")
print(f"Sample training data shape: {X_train[0].shape}")
print(f"Expected input shape: ({TIMESTEPS}, {N_FEATURES})")

# Verify input dimensions
assert X_train.shape[1] == TIMESTEPS, \
    f"Mismatched timesteps: Data has {X_train.shape[1]}, expected {TIMESTEPS}"
assert X_train.shape[2] == N_FEATURES, \
    f"Mismatched features: Data has {X_train.shape[2]}, expected {N_FEATURES}"

# --------- 3. SVR Hyperparameter Tuning ---------
svr_param_grid = {
    'C': np.logspace(-1, 3, 10),
    'epsilon': [0.01, 0.1, 0.5, 1.0],
    'kernel': ['rbf', 'poly', 'sigmoid']
}

svr_search = RandomizedSearchCV(
    SVR(),
    svr_param_grid,
    n_iter=50,
    cv=SVR_CV,
    scoring='neg_mean_squared_error',
    n_jobs=-1,
    verbose=1
)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_features)
X_test_scaled = scaler.transform(X_test_features)

svr_search.fit(X_train_scaled, y_train.ravel())

# Get best SVR
best_svr = svr_search.best_estimator_

# --------- 4. Final Ensemble Model ---------
class LSTMSVREnsemble:
    def __init__(self, lstm_model, svr_model, scaler):
        self.lstm_feature_extractor = Model(
            inputs=lstm_model.input,
            outputs=lstm_model.layers[-2].output
        )
        self.scaler = scaler
        self.svr_model = svr_model
        
    def predict(self, X):
        features = self.lstm_feature_extractor.predict(X)
        scaled_features = self.scaler.transform(features)
        return self.svr_model.predict(scaled_features)

# Initialize and evaluate final model
final_model = LSTMSVREnsemble(best_lstm, best_svr, scaler)
y_pred = final_model.predict(X_test)

# Calculate performance
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"\nFinal Ensemble RMSE: {rmse:.4f}")
print(f"Best LSTM Params: {lstm_tuner.get_best_hyperparameters()[0].values}")
print(f"Best SVR Params: {svr_search.best_params_}")
````
